## **Business Problem Statement**

### ***Case Background***

**Perusahaan XYZ** merupakan perusahaan e-commerce yang menyediakan platform transaksi online bagi pelanggan individu dengan beragam kategori produk, metode pembayaran, serta pengalaman belanja digital yang terintegrasi melalui aplikasi dan website. Pelanggan dapat dengan mudah melakukan pencarian produk, memesan barang, memanfaatkan promo, hingga melakukan pembayaran secara fleksibel sesuai preferensi masing-masing.

Sebagai perusahaan yang beroperasi di industri e-commerce yang sangat kompetitif, Perusahaan XYZ menghadapi tantangan besar dalam mempertahankan pelanggan. Konsumen e-commerce cenderung memiliki tingkat loyalitas yang rendah karena banyaknya pilihan platform dengan penawaran harga, promo, dan pengalaman pengguna yang serupa. Persaingan antar platform mendorong pelanggan untuk dengan mudah berpindah ke kompetitor ketika merasa tidak puas terhadap layanan, harga, maupun pengalaman aplikasi.

**Customer churn** atau hilangnya pelanggan menjadi salah satu tantangan utama dalam industri e-commerce. Berdasarkan data internal Perusahaan XYZ, terlihat adanya penurunan aktivitas transaksi pada segmen pelanggan yang sebelumnya aktif dan loyal. Kondisi ini mengindikasikan potensi churn yang dapat berdampak langsung pada penurunan pendapatan serta hilangnya nilai pelanggan jangka panjang (Customer Lifetime Value).

Selain itu, berbagai studi industri menunjukkan bahwa biaya untuk memperoleh pelanggan baru jauh lebih tinggi dibandingkan mempertahankan pelanggan yang sudah ada. [**Penelitian oleh Rudd et al. (2022)**](https://link.springer.com/article/10.1007/s44230-022-00006-y) menyebutkan bahwa biaya akuisisi pelanggan baru dapat mencapai **5 hingga 6 kali lipat** dibandingkan biaya retensi. Oleh karena itu, strategi retensi pelanggan menjadi semakin krusial untuk menjaga keberlanjutan bisnis di tengah meningkatnya biaya pemasaran dan akuisisi.

### ***Problem Statement***

Perusahaan XYZ saat ini **belum memiliki sistem yang mampu mengidentifikasi pelanggan yang berisiko churn secara dini**. Meskipun perusahaan memiliki data historis pelanggan yang cukup lengkap, mencakup perilaku transaksi, penggunaan aplikasi, serta tingkat kepuasan dan komplain, data tersebut belum dimanfaatkan secara optimal untuk tujuan prediktif.

Akibatnya, upaya retensi pelanggan masih bersifat reaktif dan tidak terfokus. Tim bisnis tidak dapat memprioritaskan pelanggan yang benar-benar berisiko churn, sehingga intervensi yang dilakukan sering kali terlambat atau tidak tepat sasaran. Kondisi ini berpotensi menyebabkan **hilangnya pelanggan loyal yang seharusnya masih dapat dipertahankan**, serta meningkatkan biaya akuisisi pelanggan baru yang relatif lebih mahal dibandingkan biaya retensi.

Oleh karena itu, **diperlukan sebuah pendekatan berbasis data untuk membangun model prediksi customer churn** yang mampu mendeteksi pelanggan berisiko sejak dini. Model ini diharapkan dapat membantu perusahaan dalam **memahami faktor-faktor utama penyebab churn**, **memprediksi pelanggan yang berpotensi berhenti menggunakan layanan**, serta **mendukung pengambilan keputusan bisnis yang lebih proaktif dan efisien** dalam strategi retensi pelanggan.

### ***Goals***

- **Mengidentifikasi faktor-faktor** utama yang mempengaruhi terjadinya customer churn di perusahaan XYZ
- **Mengembangkan model machine learning** berbasis klasifikasi untuk memprediksi pelanggan dengan risiko churn tinggi, sehingga perusahaan dapat melakukan intervensi proaktif, meminimalkan kehilangan pelanggan loyal, dan mendukung pengambilan keputusan bisnis yang lebih efektif.

### **Limitasi Model**

1. **Data Tidak Bersifat Time Series**

   Data yang digunakan berasal dari satu periode waktu sehingga model belum menangkap perubahan perilaku pelanggan secara dinamis dari waktu ke waktu.

2. **Ketidakseimbangan Kelas Churn**

   Jumlah pelanggan churn relatif lebih sedikit dibandingkan non-churn. Meskipun telah dilakukan penanganan *class imbalance*, kondisi ini tetap berpotensi memengaruhi performa model dalam mendeteksi churn.

3. **Fitur Bersifat Indikator Tidak Langsung**

   Variabel yang digunakan merepresentasikan pola perilaku pelanggan, namun tidak secara langsung menjelaskan penyebab churn. Oleh karena itu, hasil model perlu diinterpretasikan dengan konteks bisnis yang tepat.

4. **Definisi Churn Bersifat Prediktif**

   Churn didefinisikan berdasarkan potensi ketidakaktifan pelanggan, bukan konfirmasi berhentinya penggunaan layanan secara permanen.

5. **Nilai Ekonomi Pelanggan Belum Dipertimbangkan**

   Model memperlakukan seluruh pelanggan secara setara tanpa mempertimbangkan perbedaan nilai bisnis masing-masing pelanggan.


### ***Analytical Approach***

Untuk menyelesaikan permasalahan customer churn pada **Perusahaan XYZ**, digunakan pendekatan machine learning melalui pengembangan beberapa model klasifikasi, seperti *Logistic Regression, Random Forest,* serta model machine learning yang mengombinasikan beberapa model untuk meningkatkan akurasi prediksi, guna memprediksi pelanggan yang berisiko churn berdasarkan **pola historis perilaku dan pengalaman pelanggan**. Mengingat distribusi data churn yang tidak seimbang, dilakukan penanganan *class imbalance* agar model lebih sensitif dalam mendeteksi pelanggan churn sebagai kelas minoritas yang krusial bagi bisnis. Evaluasi performa model difokuskan pada metrik ***F2-Score***, yang memberikan penekanan lebih besar pada *recall* untuk meminimalkan *False Negative*, yaitu pelanggan churn yang tidak teridentifikasi oleh model dan berpotensi menimbulkan kerugian bisnis. Hasil pemodelan ini menjadi dasar solusi prediktif yang memungkinkan perusahaan melakukan intervensi retensi secara proaktif, sehingga strategi pengelolaan churn dapat beralih dari pendekatan reaktif menjadi preventif dan berbasis data.

### **Target Definition**

Dalam penelitian ini, customer **churn** didefinisikan sebagai **kondisi ketika pelanggan diprediksi tidak lagi melakukan transaksi dalam periode waktu tertentu pada masa mendatang.** Pelanggan dengan karakteristik tersebut diklasifikasikan sebagai **churn (1)**.
Sebaliknya, **non-churn** didefinisikan sebagai kondisi **ketika pelanggan masih diprediksi melakukan transaksi atau tetap aktif dalam periode waktu yang sama**, sehingga diklasifikasikan sebagai **non-churn (0)**.

### ***Metric Evaluation***

| **ACTUAL / PREDICTED** | **Not Churn** | **Churn** |
|--------------------------|---------------|------------|
| **Not Churn** | **True Negative (TN)**<br>Model predicts customer will not churn, and indeed does not churn. | **False Positive (FP)**<br>Model predicts customer will churn, but actually does not churn. |
| **Churn** | **False Negative (FN)**<br>Model predicts customer will not churn, but actually churns. | **True Positive (TP)**<br>Model predicts customer will churn, and indeed truly churns. |

Dalam evaluasi model prediksi customer churn, terdapat dua jenis kesalahan utama yang perlu diperhatikan, yaitu Type 1 Error dan Type 2 Error, yang masing-masing memiliki dampak bisnis yang berbeda.

- **Type 1 Error (False Positive)**

    Type 1 Error terjadi ketika model memprediksi seorang pelanggan akan churn (**positif**), padahal pada kenyataannya pelanggan tersebut tidak churn (**negatif**).

    Kesalahan ini dapat menyebabkan perusahaan memberikan insentif atau program retensi kepada pelanggan yang sebenarnya tidak berisiko churn, sehingga berpotensi menimbulkan **pemborosan sumber daya**. Namun, dari sisi bisnis, dampak finansial kesalahan ini relatif lebih kecil dibandingkan kehilangan pelanggan yang benar-benar churn.
    <br><br>

- **Type 2 Error (False Negative)**

    Type 2 Error terjadi ketika model memprediksi seorang pelanggan tidak akan churn (**negatif**), padahal sebenarnya pelanggan tersebut berisiko churn (**positif**).

    Kesalahan ini jauh lebih **kritis**, karena pelanggan yang seharusnya dapat diselamatkan justru terlewat, sehingga perusahaan kehilangan pendapatan dan harus menanggung **biaya akuisisi pelanggan baru** yang jauh lebih tinggi dibandingkan biaya retensi.
    <br><br>

**Fokus Evaluasi Model**

Berdasarkan pertimbangan bisnis tersebut, tujuan utama pengembangan model ini adalah **meminimalkan False Negative**, agar sebanyak mungkin pelanggan yang berisiko churn dapat terdeteksi secara akurat dan ditangani secara proaktif.
<br><br>

**Metrik yang Digunakan: F2-Score**

Untuk mendukung tujuan tersebut, metrik utama yang digunakan dalam evaluasi model adalah F2-Score. **F2-Score memberikan bobot yang lebih besar pada Recall**, sehingga model lebih difokuskan pada kemampuan menangkap pelanggan yang benar-benar berisiko churn, **tanpa mengabaikan Precision secara berlebihan**.

Rumus F2-Score didefinisikan sebagai berikut:

$$
F2\text{-Score} = \frac{(1 + 2^2) \cdot \text{Precision} \cdot \text{Recall}}{(2^2 \cdot \text{Precision}) + \text{Recall}}
$$

Dimana:
- **Precision** = $\frac{TP}{TP + FP}$

- **Recall** = $\frac{TP}{TP + FN}$

- **TP** = True Positives (banyaknya prediksi benar yang churn)

- **FP** = False Positives (banyaknya prediksi salah yang churn)

- **FN** = False Negatives (banyaknya prediksi salah yang tidak churn)

Dengan menggunakan **F2-Score** sebagai metrik utama, model diharapkan lebih selaras dengan kebutuhan bisnis Perusahaan XYZ, yaitu mengurangi kehilangan pelanggan bernilai tinggi melalui deteksi dini churn dan intervensi yang tepat sasaran.

#### **Profil Pelanggan dan Karakteristik Dasar**

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import MaxNLocator

sns.set(style="whitegrid")
overall_churn_rate = df_model['Churn'].mean() * 100

def plot_distribution_and_churnrate(cat_col, title):

    fig, axes = plt.subplots(1, 2, figsize=(14, 4))

    # Distribution (Count)

    order = (
    df_model[cat_col]
    .value_counts()
    .sort_values(ascending=False)
    .index
)

    sns.countplot(
        data=df_model,
        x=cat_col,
        hue='Churn',
        order=order,                   
        palette=['#1f77b4', '#ff7f0e'],
        ax=axes[0]
    )

    axes[0].set_title(f'Customer Churn Distribution by {cat_col}')
    axes[0].set_xlabel(cat_col)
    axes[0].set_ylabel('Count')

    for p in axes[0].patches:
        if p.get_height() > 0:
            axes[0].annotate(
                f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center',
                va='bottom',
                fontsize=9
            )

    # Churn Rate
    churn_rate = (
        df_model
        .groupby(cat_col)['Churn']
        .mean()
        .sort_values() * 100
    )

    axes[1].barh(
        churn_rate.index.astype(str),
        churn_rate.values,
        color='#c0392b'
    )

    axes[1].axvline(
        overall_churn_rate,
        linestyle='--',
        color='black',
        label=f'Overall Churn Rate: {overall_churn_rate:.2f}%'
    )

      
    for i, v in enumerate(churn_rate.values):
        axes[1].xaxis.set_major_locator(MaxNLocator(integer=True))
        axes[1].text(v + 0.5, i, f'{v:.2f}%', va='center')

    axes[1].set_title(f'Churn Rate by {cat_col}')
    axes[1].set_xlabel('Churn Rate (%)')
    axes[1].legend()

    plt.suptitle(title, fontsize=13, fontweight='bold')
    plt.tight_layout()
    plt.show()
```

```python
plot_distribution_and_churnrate(
    'Gender',
    'Customer Churn Analysis by Gender'
)
```

Berdasarkan visualisasi distribusi dan churn rate tersebut bisa dilihat bahwa pelanggan laki-laki memiliki churn rate sebesar 17,73%, sedikit lebih tinggi dibandingkan pelanggan perempuan sebesar 15,49% dan berada di atas rata-rata keseluruhan. Meskipun demikian, perbedaannya relatif kecil, sehingga gender bukan faktor utama churn dan lebih berperan sebagai variabel pendukung.

#### **Analisis Fitur Numerik dan Kategorikal terhadap Churn**

**Perbandingan Fitur Numerik dengan Churn**

Analisis difokuskan pada fitur numerik untuk memahami faktor-faktor yang berpotensi memengaruhi churn. Karena target (Churn) bersifat kategorikal (0 = tidak churn, 1 = churn), analisis dilakukan dengan membandingkan distribusi fitur numerik antara kedua grup.

Sebelum uji, dilakukan **uji normalitas** menggunakan *Shapiro-Wilk* untuk mengetahui apakah fitur berdistribusi normal:

- H₀ → Data numerik berdistribusi normal
- H₁ → Data numerik tidak berdistribusi normal
- p-value < 0,05 → H₀ ditolak → data tidak normal
- p-value ≥ 0,05 → H₀ gagal ditolak → data kemungkinan normal

```python
numerical_cols = [ 'Tenure', 
                  'CityTier', 
                  'DaySinceLastOrder', 
                  'CashbackAmount', 
                  'OrderCount', 
                  'OrderAmountHikeFromlastYear', 
                  'CouponUsed', 
                  'HourSpendOnApp', 
                  'SatisfactionScore', 
                  'NumberOfDeviceRegistered', 
                  'WarehouseToHome', 
                  'Complain', 
                  'NumberOfAddress']
results = []

for col in numerical_cols:
    stat, p = shapiro(df_model[col].dropna())
    
    results.append({
        'Variable': col,
        'Shapiro_Statistic': stat,
        'p_value': p,
        'Normality': 'Normal' if p >= 0.05 else 'Tidak Normal',
        'Conclusion': (
            'Gagal tolak H0 (Data berdistribusi normal)' 
            if p >= 0.05 
            else 'Tolak H0 (Data tidak berdistribusi normal)'
        )
    })

df_shapiro_result = pd.DataFrame(results)

df_shapiro_result
```

Hasil uji normalitas menunjukkan sebagian besar fitur tidak berdistribusi normal, sehingga dipilih *Mann-Whitney U test*, yaitu uji non-parametrik untuk membandingkan distribusi dua grup (churn dan non-churn):

- H₀ → Distribusi fitur sama antara churn dan non-churn
- H₁ → Distribusi fitur berbeda antara churn dan non-churn
- p-value < 0,05 → H₀ ditolak → distribusi berbeda signifikan
- p-value ≥ 0,05 → H₀ gagal ditolak → distribusi tidak berbeda signifikan

```python
mannwhitney_results = {}

for col in numerical_cols:
    group0 = df_model[df_model['Churn']==0][col]
    group1 = df_model[df_model['Churn']==1][col]
    
    stat, p = mannwhitneyu(group0, group1, alternative='two-sided')
    mannwhitney_results[col] = (stat, p)

mw_df = pd.DataFrame({
    'Feature': list(mannwhitney_results.keys()),
    'U-stat': [v[0] for v in mannwhitney_results.values()],
    'p-value': [v[1] for v in mannwhitney_results.values()],
})

# Tambahkan kolom decision berdasarkan p-value masing-masing
mw_df['decision'] = mw_df['p-value'].apply(lambda p: "Reject H₀ (Significant)" if p < 0.05 else "Fail to Reject H₀ (Not Significant)")

mw_df = mw_df.sort_values(by='p-value')

mw_df

```

**Fitur Kategorikal dengan Churn**

Setelah melakukan eksplorasi pada variabel numerik, langkah berikutnya adalah menganalisis bagaimana variabel-variabel kategorikal berhubungan dengan perilaku churn.

```python
from scipy.stats import chi2_contingency

cat_cols = df_model[['PreferredLoginDevice', 'PreferredPaymentMode', 'Gender', 'PreferedOrderCat', 'MaritalStatus']]
chi2_results = []

for col in cat_cols:
    if col != 'Churn':
        table = pd.crosstab(df_model[col], df_model['Churn'])

        chi2, p, dof, expected = chi2_contingency(table)

        chi2_results.append({
            'Feature': col,
            'Chi2 Statistic': chi2,
            'p-value': p,
            'Decision': 'Reject H₀ (Significant)' if p < 0.05 else 'Fail to Reject H₀ (Not Significant)'
        })

chi2_df = pd.DataFrame(chi2_results).sort_values(by='p-value')
chi2_df
```

Berdasarkan hasil uji Chi-Square menunjukkan bahwa setiap fitur kategorikal dari Dataset E-commerce tersebut memiliki hubungan yang signifikan secara statistik.
- H₀ (Hipotesis nol) → Tidak ada hubungan antara variabel kategorikal dan churn.
- H₁ (Hipotesis alternatif) → Ada hubungan antara variabel kategorikal dan churn.
Jika nilai p-value < 0.05, maka kita menolak H₀, artinya fitur tersebut berhubungan signifikan dengan perilaku churn pelanggan.

**Handling Data Imbalanced**

```python
# Define resampler
smote = SMOTE(random_state=0)
ros = RandomOverSampler(random_state=0)
nmiss = NearMiss(version=1)
rus = RandomUnderSampler(random_state=0)

resamplers = [smote, ros, nmiss, rus]
```

Selanjutnya, karena data yang digunakan merupakan **data imbalance**, di mana jumlah sampel pada kelas mayoritas jauh lebih besar dibanding kelas minoritas, kondisi ini dapat menyebabkan model bias terhadap kelas mayoritas dan menurunkan kemampuan model dalam mengenali kelas minoritas (churn). Untuk mengatasi hal tersebut, digunakan beberapa teknik *resampling* sebagai berikut:

1. **SMOTE (Synthetic Minority Over-sampling Technique)** : teknik *oversampling* yang menghasilkan contoh baru pada kelas minoritas melalui proses interpolasi antar sampel yang berdekatan. Metode ini meningkatkan proporsi kelas minoritas tanpa sekadar menduplikasi data. Teknik ini digunakan karena mampu menangkap variasi data minoritas secara lebih baik dan sering memberikan peningkatan performa model pada data yang sangat tidak seimbang.

2. **Random Oversampler** : teknik *oversampling* sederhana yang bekerja dengan menggandakan sampel kelas minoritas secara acak hingga proporsinya lebih seimbang. Metode ini digunakan karena memberikan baseline yang mudah dipahami, cepat diimplementasikan, serta membantu meningkatkan sensitivitas model terutama pada kelas minoritas.

3. **Random UnderSampler** : teknik *undersampling* yang menyeimbangkan data dengan mengurangi jumlah sampel kelas mayoritas secara acak. Teknik ini digunakan karena dapat mempercepat proses pelatihan, mengurangi dominasi kelas mayoritas, serta membantu mengevaluasi performa model ketika data dibuat lebih ringkas.

4. **NearMiss** : teknik *undersampling* yang memilih sampel kelas mayoritas berdasarkan jarak terdekat terhadap sampel kelas minoritas. Dengan mempertahankan sampel yang paling relevan secara spasial, teknik ini membantu model membentuk batas keputusan yang lebih jelas. NearMiss digunakan karena mampu menjaga struktur hubungan antar kelas, terutama saat perbedaan distribusi terlalu besar.

#### **Model Benchmarking**

Setiap kombinasi model dan teknik resampling dievaluasi melalui pipeline yang mencakup preprocessing, resampling, dan modeling.

- Cross-validation menggunakan StratifiedKFold memastikan proporsi kelas tetap seimbang pada setiap fold.
- Evaluasi dilakukan menggunakan F2-score, sehingga model yang dipilih memiliki kemampuan mendeteksi churn secara konsisten.
- Hasil cross-validation dicatat dalam rata-rata F2 (mean), deviasi standar (std), dan performa per fold, yang digunakan untuk memilih model dengan performansi terbaik dan variasi terkecil.


Pendekatan ini menjamin pemilihan model akhir tidak hanya berdasarkan akurasi keseluruhan, tetapi juga kemampuan menangkap pelanggan churn secara andal, sesuai prioritas bisnis.

#### **Cross Validation**

Cross-Validation adalah teknik evaluasi model machine learning di mana data training dibagi menjadi beberapa bagian (fold). Model dilatih pada sebagian fold dan diuji pada fold yang tersisa secara bergantian, sehingga setiap fold pernah menjadi data uji satu kali. Pendekatan ini memberikan estimasi performa model yang lebih stabil dan akurat, membantu mendeteksi overfitting, dan mempermudah pemilihan model atau hyperparameter sebelum diuji pada data test yang benar-benar baru.

```python
f2_mean = []
f2_std = []
all_f2 = []
model_name = []
resampler_name = []

# Mencari model terbaik berdasarkan rerata f2 score terkecil dan std terkecil
for algoritma in models:
    for resampler in resamplers:

        model_pipe = ImbPipeline([
            ('preprocessing', prepros), 
            ('resampler', resampler),
            ('model', algoritma)
        ])

        skfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

        model_cv = cross_val_score(
            model_pipe,                 
            xtrain,                    
            ytrain, 
            cv = skfold, 
            scoring = f2_scorer,
            error_score='raise'
            )

        f2_mean.append(model_cv.mean())
        f2_std.append(model_cv.std())
        all_f2.append(model_cv.round(4))
        model_name.append(algoritma.__class__.__name__)
        resampler_name.append(resampler.__class__.__name__)
```

```python
cv_results_df = pd.DataFrame({
    'model': model_name,
    'resampler': resampler_name,
    'f2 (cv)': f2_mean,
    'std': f2_std
}).sort_values(['f2 (cv)','std'], ascending=False)
cv_results_df.head(10)
```

#### **Test Set Evaluation**

Setelah benchmarking menggunakan cross-validation, seluruh model diuji pada test set yang belum pernah dilihat sebelumnya. Evaluasi dilakukan menggunakan F2-score, sesuai prioritas bisnis yang menekankan recall tinggi agar pelanggan churn yang berpotensi terdeteksi dengan maksimal.

```python
# Evaluate the models on the test set
test_f2_scores = []
model_names_test = []
resampler_names_test = []

for algoritma in models:
    for resampler in resamplers:
        model_pipe = ImbPipeline([
            ('preprocessor', prepros),
            ('resampler', resampler),
            ('model', algoritma)
        ])

        # Fitting the model
        model_pipe.fit(xtrain, ytrain)

        # Predicting
        y_pred = model_pipe.predict(xtest)

        # Evaluate with F2 score
        test_f2 = fbeta_score(ytest, y_pred, beta=2)
        test_f2_scores.append(test_f2)
        model_names_test.append(algoritma.__class__.__name__)
        resampler_names_test.append(resampler.__class__.__name__)
```

```python
# Store test set results in a DataFrame
test_results_df = pd.DataFrame({
    'model': model_names_test,
    'resampler': resampler_names_test,
    'f2 (test_set)': test_f2_scores
}).sort_values('f2 (test_set)', ascending=False)


test_results_df.head(10)
```

```python
merged_df = pd.merge(cv_results_df, test_results_df, on=['model', 'resampler']).drop(columns=['std'])
merged_df['diff'] = merged_df['f2 (test_set)'] - merged_df['f2 (cv)']
pd.set_option('display.precision', 10)

merged_df.sort_values(['f2 (test_set)'], ascending=False).head(10)
```

### **Model Evaluation**

Setelah melakukan data training/modelling, maka selanjutnya adalah dilakukan model evaluation berdasarkan best model yang telah dipilih melalui proses benchmarking. Pada tahap ini, model terbaik yang sudah dipilih sebelumnya kembali diuji secara lebih mendalam untuk memastikan bahwa performanya benar-benar optimal. Pertama, model dicoba apa adanya untuk melihat kemampuan dasarnya. Setelah itu, model disempurnakan dengan menyesuaikan parameternya agar bisa bekerja lebih akurat. Model yang sudah disempurnakan kemudian diuji kembali menggunakan data yang benar-benar baru, sehingga dapat dilihat apakah model mampu mengenali pola churn secara konsisten dan dapat diandalkan. Hasil akhirnya dibandingkan untuk memastikan bahwa model final yang dipilih adalah yang memberikan performa paling baik dan stabil untuk kebutuhan prediksi churn.

#### **1. XGBoost + SMOTE**

```python
# Define hyperparameter XGBoost
xgb_param_grid = {
    'model__n_estimators': [200, 300],
    'model__max_depth': [4, 6, 8],
    'model__learning_rate': [0.05, 0.1],
    'model__subsample': [0.8, 1.0],
    'model__colsample_bytree': [0.8, 1.0]
}

# Create pipeline
xgb_smote_pipe = ImbPipeline([
    ('preprocessing', prepros),
    ('resampler', smote),
    ('model', xgboost)
])

# Untuned XGB
xgb_smote_untuned = xgb_smote_pipe.fit(xtrain, ytrain)
y_pred_xgb_smote_untuned = xgb_smote_untuned.predict(xtest)

xgb_smote_untuned_f2 = fbeta_score(ytest, y_pred_xgb_smote_untuned, beta=2)
xgb_smote_untuned_recall = recall_score(ytest, y_pred_xgb_smote_untuned)
xgb_smote_untuned_precision = precision_score(ytest, y_pred_xgb_smote_untuned)

# Tuned XGB dengan GridSearch
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

xgb_smote_grid = GridSearchCV(
    estimator=xgb_smote_pipe,
    param_grid=xgb_param_grid,
    scoring=f2_scorer,
    cv=cv,
    n_jobs=-1,
    verbose=1
)

xgb_smote_grid.fit(xtrain, ytrain)

xgb_smote_best = xgb_smote_grid.best_estimator_
y_pred_xgb_smote_tuned = xgb_smote_best.predict(xtest)


best_index = xgb_smote_grid.best_index_
xgb_smote_cv_std = xgb_smote_grid.cv_results_['std_test_score'][best_index]
xgb_smote_tuned_f2 = fbeta_score(ytest, y_pred_xgb_smote_tuned, beta=2)
xgb_smote_tuned_recall = recall_score(ytest, y_pred_xgb_smote_tuned)
xgb_smote_tuned_precision = precision_score(ytest, y_pred_xgb_smote_tuned)

print("XGB best params:", xgb_smote_grid.best_params_)
print("XGB best F2 (CV mean):", xgb_smote_grid.best_score_)
print("XGB F2 (CV std):", xgb_smote_cv_std)
print("XGB F2 (test, tuned):", xgb_smote_tuned_f2)
```

```python
# Evaluate untuned model on train set
y_pred_train_untuned_xgb_smote = xgb_smote_untuned.predict(xtrain)
train_untuned_f2_xgb_smote = fbeta_score(ytrain, y_pred_train_untuned_xgb_smote, beta=2)
train_untuned_recall_xgb_smote = recall_score(ytrain, y_pred_train_untuned_xgb_smote)
train_untuned_precision_xgb_smote = precision_score(ytrain, y_pred_train_untuned_xgb_smote)

# Evaluate tuned model on train set
y_pred_train_tuned_xgb_smote = xgb_smote_best.predict(xtrain)
train_tuned_f2_xgb_smote = fbeta_score(ytrain, y_pred_train_tuned_xgb_smote, beta=2)
train_tuned_recall_xgb_smote = recall_score(ytrain, y_pred_train_tuned_xgb_smote)
train_tuned_precision_xgb_smote = precision_score(ytrain, y_pred_train_tuned_xgb_smote)

# Buat perbandingan
comparison_xgb = pd.DataFrame({
    'Model': ['Untuned XGB (SMOTE)', 'Tuned XGB (SMOTE)'],
    'F2 Score (Train Set)': [train_untuned_f2_xgb_smote, train_tuned_f2_xgb_smote],
    'F2 Score (Test Set)': [xgb_smote_untuned_f2, xgb_smote_tuned_f2],
    'F2 CV Std Dev': [np.nan, xgb_smote_cv_std],
    'Recall (Train Set)': [train_untuned_recall_xgb_smote, train_tuned_recall_xgb_smote],
    'Recall (Test Set)': [xgb_smote_untuned_recall, xgb_smote_tuned_recall],
    'Precision (Train Set)': [train_untuned_precision_xgb_smote, train_tuned_precision_xgb_smote],
    'Precision (Test Set)': [xgb_smote_untuned_precision, xgb_smote_tuned_precision]
})


comparison_xgb
```

```python
# Define hyperparameter LightGBM
lgbm_param_grid = {
    'model__n_estimators': [200, 300],
    'model__max_depth': [-1, 6, 8],
    'model__learning_rate': [0.05, 0.1],
    'model__num_leaves': [31, 63],
    'model__subsample': [0.8, 1.0]
}

# Create pipeline LGBM + RandomOverSampler
lgbm_pipe = ImbPipeline([
    ('preprocessing', prepros),
    ('resampler', ros),
    ('model', lgbm)
])

# Untuned LGBM
lgbm_untuned = lgbm_pipe.fit(xtrain, ytrain)
y_pred_lgbm_untuned = lgbm_untuned.predict(xtest)

lgbm_untuned_f2 = fbeta_score(ytest, y_pred_lgbm_untuned, beta=2)
lgbm_untuned_recall = recall_score(ytest, y_pred_lgbm_untuned)
lgbm_untuned_precision = precision_score(ytest, y_pred_lgbm_untuned)

# Tuned LGBM dengan GridSearch
lgbm_grid = GridSearchCV(
    estimator=lgbm_pipe,
    param_grid=lgbm_param_grid,
    scoring=f2_scorer,
    cv=cv,
    n_jobs=-1,
    verbose=1
)

lgbm_grid.fit(xtrain, ytrain)

lgbm_best = lgbm_grid.best_estimator_
y_pred_lgbm_tuned = lgbm_best.predict(xtest)


best_index_lgbm = lgbm_grid.best_index_
lgbm_cv_std = lgbm_grid.cv_results_['std_test_score'][best_index_lgbm]
lgbm_tuned_f2 = fbeta_score(ytest, y_pred_lgbm_tuned, beta=2)
lgbm_tuned_recall = recall_score(ytest, y_pred_lgbm_tuned)
lgbm_tuned_precision = precision_score(ytest, y_pred_lgbm_tuned)
```

### **Recommendations**

**1. Perkuat Retensi Pelanggan Baru & Early Tenure**

Karena **RS__Tenure** menjadi faktor terkuat dan pelanggan dengan masa bergabung pendek paling rentan churn, perusahaan perlu:

- Menjalankan **welcome journey** yang terstruktur (edukasi fitur, benefit, promo awal).
- Memberikan **insentif khusus untuk transaksi kedua dan ketiga** agar pelanggan tidak berhenti di one-time purchase.
- Membuat **alert internal** untuk pelanggan dengan tenure pendek yang mulai jarang bertransaksi (DaySinceLastOrder meningkat).



**2. Prioritaskan Penanganan Pelanggan yang Pernah Complain**

Fitur **RS__Complain** muncul sebagai salah satu pendorong churn terbesar, sehingga:

- Terapkan **SLA penanganan komplain yang ketat** dan pantau hingga benar-benar terselesaikan.
- Tag pelanggan yang pernah komplain sebagai **“high-risk segment”** dan berikan **follow-up personal** (voucher permintaan maaf, survey penyelesaian masalah).
- Integrasikan data komplain ke dalam **kampanye retensi**, sehingga pelanggan yang komplain tidak dibiarkan pasif terlalu lama.

- Luncurkan **kampanye reaktivasi** (push notif, email, in-app message) ketika DaySinceLastOrder melewati ambang tertentu.
- Gunakan pola ini untuk membangun **churn watchlist** yang dipantau berkala oleh tim CRM/marketing.



**4. Optimalkan Strategi Cashback & Value Proposition**

- Desain **skema cashback bertingkat** untuk pelanggan yang transaksinya mulai menurun, bukan hanya untuk akuisisi awal.
- Monitor pelanggan yang belanjanya turun signifikan dari periode sebelumnya dan target dengan **personalized offer** (cashback ekstra / bundle menarik).

- Sesuaikan **janji layanan (delivery promise)** berdasarkan CityTier agar ekspektasi pelanggan lebih realistis.
